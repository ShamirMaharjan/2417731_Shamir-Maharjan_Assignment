1. Create a new column called Happiness Category that categorizes countries into three categories
based on their Score:

Low − (Score < 4)
Medium − (4 ≤ Score ≤ 6)
High − (Score > 6)

```python
# Define a function for categorization
def categorize_happiness(score):
    if (score < 4):
        return 'Low'
    elif(4 <= score <= 6):
        return 'Medium'
    else:
        return 'High'

df['Happiness_Category'] = df['score'].apply(categorize_happiness)
print(df)
```

Task - 3 - Outlier Detection:

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
def find_outliers(data,column):
    Q1 = np.percentile(data[column],25); #first quartile 
    Q3 = np.percentile(data[column],75); #third quartile 
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR 
    # print(lower_bound,upper_bound)

    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    
```

2. Define outliers using the 1.5 × IQR rule.

```python
print("Score outliers: ")
score_outliers = find_outliers(south_asia_df,'score')
print(score_outliers)
print("gdp_outliers: ")
gdp_outliers =  find_outliers(south_asia_df,'Log GDP per capita')
print(gdp_outliers)
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python
outlier_countries = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(south_asia_df['Log GDP per capita'],south_asia_df['score'], color='skyblue', label='countries')
plt.scatter(outlier_countries['Log GDP per capita'],outlier_countries['score'], color='red', label='outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot: GDP per Capita vs Score (Outliers Highlighted)')
plt.legend()
plt.grid(True)
plt.show()
```

4. Discuss the characteristics of these outliers and their potential impact on regional averages.

From the above graph we can clearly see that Afghanistan stands out with a very low score (1.721) and Log GDP per Capita (0.628), reflecting its significant deviation from regional averages. These types of outliers can heavily skew regional averages downward, concealing the relative performance of other countries.

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

```python
def find_outliers(data,column):
    Q1 = np.percentile(data[column],25); #first quartile 
    Q3 = np.percentile(data[column],75); #third quartile 
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR 
    # print(lower_bound,upper_bound)

    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    
```

```python
print("South Asia Score outliers : ")
sa_score_outliers = find_outliers(south_asia_df,'score')
print(sa_score_outliers)
print("South Asia gdp_outliers: ")
sa_gdp_outliers =  find_outliers(south_asia_df,'Log GDP per capita')
print(sa_gdp_outliers)
```

```python
print("Middle East Score outliers : ")
me_score_outliers = find_outliers(middle_east_countries_df,'score')
print(me_score_outliers)
print("Middle East gdp_outliers: ")
me_gdp_outliers =  find_outliers(middle_east_countries_df,'Log GDP per capita')
print(me_gdp_outliers)
```

```python
south_asia_combined_outliers = pd.concat([sa_score_outliers, sa_gdp_outliers]).drop_duplicates()
middle_east_combined_outliers = pd.concat([me_score_outliers, me_gdp_outliers]).drop_duplicates()
```

```python
plt.figure(figsize=(14, 7))

# South Asia outliers
plt.subplot(1, 2, 1)
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], color='blue', label='South Asia Data')
plt.scatter(sa_score_outliers['Log GDP per capita'], sa_score_outliers['score'], color='red', label='Score Outliers')
plt.scatter(sa_gdp_outliers['Log GDP per capita'], sa_gdp_outliers['score'], color='green', label='GDP Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('South Asia Outliers')
plt.legend()
plt.grid(True)

# Middle East outliers
plt.subplot(1, 2, 2)
plt.scatter(middle_east_countries_df['Log GDP per capita'], middle_east_countries_df['score'], color='blue', label='Middle East Data')
plt.scatter(me_score_outliers['Log GDP per capita'], me_score_outliers['score'], color='red', label='Score Outliers')
plt.scatter(me_gdp_outliers['Log GDP per capita'], me_gdp_outliers['score'], color='green', label='GDP Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Middle East Outliers')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Discussion
print("South Asia Outliers:")
print(south_asia_combined_outliers)
print("\nMiddle East Outliers:")
print(middle_east_combined_outliers)
```
